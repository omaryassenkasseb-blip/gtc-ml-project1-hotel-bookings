```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Upload CSV file
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv('hotel_bookings.csv')

# Preview data
print(df.head())
```

```python
# Display column names
print("Column names:")
print(df.columns)

# Drop duplicates
df.drop_duplicates(inplace=True)

# Convert date column
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Fill missing values
df['children'].fillna(0, inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Create new features
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_family'] = (df['children'] + df['babies']) > 0

# Check if 'is_canceled' column exists
if 'is_canceled' in df.columns:
    sns.countplot(x='is_canceled', data=df, palette='Set2')
    plt.title('Booking Cancellation Rate')
    plt.xlabel('Canceled (0 = No, 1 = Yes)')
    plt.ylabel('Number of Bookings')
    plt.show()
else:
    print("Column 'is_canceled' not found in the dataset.")
```

```python
# Hotel type distribution
sns.countplot(x='hotel', data=df, palette='Set1')
plt.title('Hotel Type Distribution')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python

# Stay duration distribution
sns.histplot(df['total_stay'], bins=30, kde=True, color='skyblue')
plt.title('Stay Duration Distribution')
plt.xlabel('Number of Nights')
plt.ylabel('Number of Bookings')
plt.show()
```

```python


# Family bookings
sns.countplot(x='is_family', data=df, palette='coolwarm')
plt.title('Family Booking Indicator')
plt.xlabel('Is Family Booking')
plt.ylabel('Number of Bookings')
plt.show()
```

```python

df['arrival_date_month'] = df['arrival_date_month'].astype(str)
sns.countplot(x='arrival_date_month', data=df, order=[
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])
plt.title('Monthly Arrival Distribution')
plt.xlabel('Arrival Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```
